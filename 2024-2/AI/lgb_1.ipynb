```python
dwn_url='https://drive.google.com/uc?id=1kIp6nFxCdnN2o4GAVcQVKQTzr7JONqac'
df = pd.read_csv(dwn_url)
```

```python
# check nan
df.isna().sum()
```

```python
# convert nan to -1
df.fillna(-1, inplace=True)
```

```python
df.info()
```

```python
# find type of the column name of df is column
for col in df.columns:
    if df[col].dtype == 'object':
        print(col)

```

```python
# convert object to float
# and target to or int
#df[target] = df[target].astype(int)
# del df object column
del df['ID']
```

```python
# target: 飆股, souring stock (0 or 1)
# use lightgbm to train model
target='飆股'
```

```python
# use lightgbm to train model
from lightgbm import LGBMClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

```

```python
# splite to train/test
X_train, X_test, y_train, y_test = train_test_split(df.drop(target, axis=1), df[target], test_size=0.3, random_state=42)
```

```python
# train model
model = LGBMClassifier()
model.fit(X_train, y_train)
```

```python
# chech test and print metric of test predict result
y_pred = model.predict(X_test)
print(accuracy_score(y_test, y_pred))
```
