<big>
Models the relationship between input variables and a continuous target variable using a straight line.</big>


\begin{align}
\text{Linear}&:& \hat y &=& a+bx\\
\text{Polynomial}&:& \hat y &=& a+\sum_{i=1}^n b_ix^i\\
\end{align}

  - Lasso Regression
<img src="https://drive.google.com/uc?export=view&id=1ppdG8GHrsD2PARtc6_kkEpDpB7KrC88X" width=600 />

<big>Regularized linear model that uses L1 (L2) penalty to perform variable selection by shrinking some coefficients to zero.
 </big>


\begin{align}
\text{Lasso}&:& \sum_i (y_i- \hat y)^2 &=& \sum_i\left( y_i-\sum_j w_j\cdot x_{ij}\right)^2+\lambda\sum_j|w_j|\\
\text{Ridge}&:&  \sum_i (y_i- \hat y)^2 &=& \sum_i\left( y_i-\sum_j w_j\cdot x_{ij}\right)^2+\lambda\sum_j w_j^2\\
\end{align}

  - Elastic Net Regression
   
<img src="https://drive.google.com/uc?export=view&id=1YlSRIBsiK1YAEaRIUy6D1PWbtNIqYvz0" width=600 />

<big>Combines L1 and L2 regularization to balance between Lasso and Ridge, especially effective when there are multiple correlated features.
 </big>

```python
def load_data_in_chunks(file_path, chunk_size=100000):
    """
    Load data from a CSV file in chunks.

    Parameters:
    - file_path: str, path to the CSV file
    - chunk_size: int, number of rows per chunk

    Returns:
    - generator of DataFrames
    """
    df = pd.DataFrame()
    for chunk in pd.read_csv(file_path, chunksize=chunk_size):
        df = pd.concat([df, chunk], ignore_index=True)
        del chunk
    return df
```

```python
# binding features
def bin_feature(x,n_bins=100):
    col_min = x.min()
    col_max = x.max()
    bins = np.linspace(col_min, col_max, n_bins + 1)
    bins_=np.digitize(x, bins[1:-1])
    return bins_
```

```python
def EDA(df_):
    """
    Perform exploratory data analysis on the DataFrame.

    Parameters:
    - df: DataFrame to analyze
    """
    all_features = len(df_.columns)
    for feature in df_.columns[:-1]:
        df_[feature]=bin_feature(df_[feature],n_bins=100)
    # --- 技術指標特徵 ---
    df_ = TI_new(df_)
    TI_size=len(df_.columns)-all_features
    print("(TI_new) New Features: ",TI_size)
    # --- 外資券商特徵 ---
    df_ = derive_broker_features1(df_, n=3)
    foreign_size=len(df_.columns)-all_features-TI_size
    print("(derive_broker_features1) New Features: ",foreign_size)
    # --- 法人與財務交互 ---
    df_ = derive_institutional_financial_features(df_)
    institutional_size=len(df_.columns)-all_features-TI_size-foreign_size
    print("(derive_institutional_financial_features) New Features: ",institutional_size)
    # --- 月營收趨勢與雙重成長邏輯 ---
    df_ = derive_monthly_revenue_features(df_)
    monthly_revenue_size=len(df_.columns)-all_features-TI_size-institutional_size-foreign_size
    print("(derive_monthly_revenue_features) New Features: ",monthly_revenue_size)
    # --- 主力籌碼異常行為與集中度 ---
    df_= derive_chips_features(df_)
    chips_size=len(df_.columns)-all_features-TI_size-institutional_size-foreign_size-monthly_revenue_size
    print("(derive_chips_features) New Features: ",chips_size)


    return df_
```
