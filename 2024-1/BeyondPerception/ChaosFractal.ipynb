import cv2
import numpy as np
import plotly.graph_objects as go
from scipy.stats import norm

# Step 1: Load the image and convert to grayscale (luminance)
def calculate_luminance(image):
    return 0.2126 * image[:, :, 0] + 0.7152 * image[:, :, 1] + 0.0722 * image[:, :, 2]

# Step 2: Calculate luminance differences for a given pixel separation R
def luminance_differences(luminance, R):
    rows, cols = luminance.shape
    diffs = []
    for i in range(rows - R):
        for j in range(cols - R):
            diff = luminance[i + R, j + R] - luminance[i, j]
            diffs.append(diff)
    return np.array(diffs)

# Step 3: Plot the final PDF for different pixel separations in one plot using Plotly
def plot_pdf_with_plotly(diffs_dict):
    fig = go.Figure()

    # Step 4: Loop through each R and add a trace (curve) to the plot
    for R, diffs in diffs_dict.items():
        hist, bins = np.histogram(diffs, bins=100, density=True)
        bin_centers = 0.5 * (bins[1:] + bins[:-1])
        
        # Fit the data to a normal distribution
        mu, std = norm.fit(diffs)
        fitted_pdf = norm.pdf(bin_centers, mu, std)

        # Add the histogram (actual PDF)
        fig.add_trace(go.Scatter(x=bin_centers, y=hist, mode='lines', name=f'R={R}', line=dict(width=2)))
        
        # Add the fitted PDF for comparison
        fig.add_trace(go.Scatter(x=bin_centers, y=fitted_pdf, mode='lines', name=f'Fit R={R}', line=dict(dash='dash')))

    # Update the layout for better visualization
    fig.update_layout(
        title='PDF of Luminance Differences for Various Pixel Separations',
        xaxis_title='Luminance Difference',
        yaxis_title='Probability Density',
        yaxis_type="log",  # Set y-axis to log scale
        legend_title='Pixel Separation (R)',
        width=900,
        height=600
    )
    
    fig.show()

# Main function
def main(image_path):
    # Load image
    image = cv2.imread(image_path)
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    # Convert to luminance
    luminance = calculate_luminance(image_rgb)
    
    # Step 5: Compute luminance differences for different pixel separations R
    diffs_dict = {}
    for R in [60, 240, 400, 600, 800, 1200]:
        diffs = luminance_differences(luminance, R)
        diffs_dict[R] = diffs
    
    # Plot the PDF using Plotly
    plot_pdf_with_plotly(diffs_dict)

# Example usage
image_path = 'path_to_your_image.png'  # Replace with the path to your image
main(image_path)

def simulate_deer_vision(image_array):
    # Convert to float for calculations
    img_float = image_array.astype(float) / 255.0
    
    # Extract RGB channels
    r, g, b = img_float[:, :, 0], img_float[:, :, 1], img_float[:, :, 2]
    
    # Simulate dichromatic vision (blue and yellow)
    # These weights are approximations based on general understanding of deer vision
    yellow = 0.5 * r + 0.5 * g  # Combining red and green to simulate yellow sensitivity
    
    # Create new image array with blue and yellow channels
    deer_vision = np.zeros_like(img_float)
    deer_vision[:, :, 0] = yellow  # Use yellow for red channel
    deer_vision[:, :, 1] = yellow  # Use yellow for green channel
    deer_vision[:, :, 2] = b       # Keep blue channel
    
    # Convert back to 8-bit format
    deer_vision = (deer_vision * 255).astype(np.uint8)
    
    return deer_vision

# Function to load and process image
def process_image(image_path):
    # Load image
    img = Image.open(image_path)
    
    # Convert to numpy array
    img_array = np.array(img)
    
    # Simulate deer vision
    deer_vision_array = simulate_deer_vision(img_array)
    
    return img_array, deer_vision_array

import numpy as np
import matplotlib.pyplot as plt

class Boids:
    def __init__(self, Np, Na, home=[0,0]):
        self.Np = Np  # predator
        self.Na = Na  # agent
        self.d = len(home)
        self.home = np.array(home)
        self.predators = np.random.random((Np, self.d*2))
        self.agents = np.random.random((Na, self.d*2))
        self.radius_repulsion = 1
        self.radius_alignment = 0.1
        
    def _pairwise_distances(self, X, Y):
        D = -2 * X @ Y.T + np.sum(Y ** 2, axis=1) + np.sum(X ** 2, axis=1)[:, np.newaxis]
        D[D < 0] = 0
        return D
        
    def _force(self):
        Daa = self._pairwise_distances(self.agents[:,:self.d], self.agents[:,:self.d])
        Dap = self._pairwise_distances(self.agents[:,:self.d], self.predators[:,:self.d])

        Neibors_repulsion = (Daa < self.radius_repulsion).astype(int)
        Neibors_alignment = (Daa < self.radius_alignment).astype(int)
        
        Dr = (1/(Daa+np.eye(self.Na))) * Neibors_repulsion
        Fr = np.diag(np.sum(Dr, axis=1))@ self.agents[:,:self.d] - Dr @ self.agents[:,:self.d]

        Fh = self.home - self.agents[:,:self.d]
    
        Fa = Neibors_alignment @ self.agents[:,self.d:] - 2*self.agents[:,self.d:]
        
        s = 100
        Ff = -np.diag(np.sqrt(np.sum(np.square(self.agents[:,self.d:]), axis=1))- s)/s @ self.agents[:,self.d:]
        
        F_total = Fr + 5*Fh + Ff + Fa
        return 200 * np.tanh(0.1*F_total)
        
    def evolve(self, dt = 0.01):
        self.agents[:,self.d:] += dt*self._force()
        self.agents[:,:self.d] += dt*self.agents[:,self.d:]
        
Na = 30
d = 2
n_history = 10
film = np.zeros((Na, d*n_history))
boids = Boids(0, Na)    
for t in range(100):        
    boids.evolve()
    film = np.hstack([boids.agents[:,:d], film[:,:-d]])
    if t > n_history:
        plt.figure()
        for i in range(Na):
            plt.plot(film[i,0],film[i,1], 'ko')
            plt.plot(film[i,::2],film[i,1::2], 'k-',alpha=0.5)
        bound = 3
        plt.xlim([-bound,bound])
        plt.ylim([-bound,bound])
        plt.title('t={}'.format(t))
        plt.savefig('./fig/{}.jpg'.format(t))

````python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Load the image
img_file= "https://drive.google.com/file/d/1vvFf7_49MKgjk6WFqkHQ0cM4QU2A56a3/view?usp=drive_link"
#img_file="https://drive.google.com/uc?export=view&id=1vvFf7_49MKgjk6WFqkHQ0cM4QU2A56a3"

#img = cv2.imread("vangogh2.jpg")
img = cv2.imread("Scream.jpg")
# Convert the image to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Apply a fractal analysis using the box-counting method
def box_counting(gray):
    # Calculate the number of boxes needed to cover the image
    boxes = np.zeros((gray.shape[0], gray.shape[1]))
    for i in range(gray.shape[0]):
        for j in range(gray.shape[1]):
            if gray[i, j] > 0:
                boxes[i, j] = 1
    # Calculate the number of boxes that contain at least one pixel
    num_boxes = np.sum(boxes)
    # Calculate the box-counting dimension
    D = np.log(num_boxes) / np.log(2)
    return D

# Apply the fractal analysis
D = box_counting(gray)

# Visualize the results
plt.imshow(gray, cmap='gray')
plt.title('Fractal Analysis of Starry Night')
plt.xlabel('Box-counting dimension: {:.2f}'.format(D))
plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage
from skimage import io, color

def calculate_turbulence(image_path):
    # Load the image and convert to grayscale
    img = io.imread(image_path)
    gray = color.rgb2gray(img)

    # Calculate gradients
    dx, dy = np.gradient(gray)

    # Calculate gradient magnitude
    magnitude = np.sqrt(dx**2 + dy**2)

    # Calculate curl (vorticity)
    curl = ndimage.sobel(dy, axis=0) - ndimage.sobel(dx, axis=1)

    # Calculate turbulence metric (you might want to adjust this)
    turbulence = np.mean(np.abs(curl) * magnitude)

    return turbulence, magnitude, curl

def plot_results(img, magnitude, curl):
    fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15, 5))

    ax1.imshow(img)
    ax1.set_title('Original Image')
    ax1.axis('off')

    ax2.imshow(magnitude, cmap='hot')
    ax2.set_title('Gradient Magnitude')
    ax2.axis('off')

    ax3.imshow(curl, cmap='coolwarm')
    ax3.set_title('Curl (Vorticity)')
    ax3.axis('off')

    plt.tight_layout()
    plt.show()

# Usage
#image_path = 'path_to_your_image.jpg'
imgf = 'Scream.jpg'
#imgf= "vincent.webp"
turbulence, magnitude, curl = calculate_turbulence(imgf)
print(f"Turbulence metric: {turbulence}")

# Plot results
img = plt.imread(imgf)
plot_results(img, magnitude, curl)
````

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage
from skimage import io, color

def load_image(file_path):
    img = io.imread(file_path)
    return color.rgb2gray(img)

def calculate_structure_function(image, max_r):
    h, w = image.shape
    sf = np.zeros(max_r)
    for r in range(1, max_r):
        shifted = np.roll(image, r, axis=1)
        diff = (image - shifted) ** 2
        sf[r] = np.mean(diff)
    return sf

def estimate_hurst_exponent(sf):
    r = np.arange(1, len(sf))
    log_sf = np.log(sf[1:])
    log_r = np.log(r)
    slope, _ = np.polyfit(log_r, log_sf, 1)
    return slope / 2

def analyze_turbulence(file_path, max_r=100):
    img = load_image(file_path)
    sf = calculate_structure_function(img, max_r)
    hurst = estimate_hurst_exponent(sf)

    plt.figure(figsize=(12, 4))

    plt.subplot(121)
    plt.imshow(img, cmap='gray')
    plt.title('Original Image')
    plt.axis('off')

    plt.subplot(122)
    plt.loglog(range(1, len(sf)), sf[1:])
    plt.title(f'Structure Function (Hurst Exponent: {hurst:.2f})')
    plt.xlabel('r')
    plt.ylabel('S(r)')

    plt.tight_layout()
    plt.show()

    return hurst

# Usage
#imgf = 'Scream.jpg'
#imgf= "vincent.webp"
file_path = 'vincent.webp'
hurst_exponent = analyze_turbulence(file_path)
print(f"Estimated Hurst Exponent: {hurst_exponent:.2f}")
````

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage
from skimage import io, color, transform

def load_and_preprocess(file_path, max_size=None):
    img = io.imread(file_path)
    gray = color.rgb2gray(img)
    if max_size and max(gray.shape) > max_size:
        scale = max_size / max(gray.shape)
        gray = transform.rescale(gray, scale, anti_aliasing=True)
    return gray

def calculate_structure_function(image, max_r):
    h, w = image.shape
    sf = np.zeros(max_r)
    for r in range(1, max_r):
        shifted = np.roll(image, r, axis=1)
        diff = (image - shifted) ** 2
        sf[r] = np.mean(diff)
    return sf

def estimate_hurst_exponent(sf):
    r = np.arange(1, len(sf))
    log_sf = np.log(sf[1:])
    log_r = np.log(r)
    slope, _ = np.polyfit(log_r, log_sf, 1)
    return slope / 2

def analyze_at_multiple_scales(file_path, scales=[1000, 2000, 4000], max_r=500):
    results = []
    for scale in scales:
        img = load_and_preprocess(file_path, max_size=scale)
        sf = calculate_structure_function(img, max_r)
        hurst = estimate_hurst_exponent(sf)
        results.append((scale, hurst))

    plt.figure(figsize=(12, 6))

    plt.subplot(121)
    img = load_and_preprocess(file_path, max_size=2000)  # for display
    plt.imshow(img, cmap='gray')
    plt.title('Original Image')
    plt.axis('off')

    plt.subplot(122)
    for scale, hurst in results:
        plt.scatter(scale, hurst, label=f'Scale {scale}')
    plt.xlabel('Image Scale')
    plt.ylabel('Hurst Exponent')
    plt.title('Hurst Exponent at Different Scales')
    plt.legend()

    plt.tight_layout()
    plt.show()

    return results

# Usage
file_path = 'vangogh-2.webp'
results = analyze_at_multiple_scales(file_path)
for scale, hurst in results:
    print(f"Scale: {scale}, Hurst Exponent: {hurst:.2f}")
````

````python
from scipy.stats import linregress
import numpy as np

def fractal_dimension(luminance):
    # Calculate the box-counting dimension
    box_sizes = np.logspace(0, 3, 10)
    box_counts = np.zeros(len(box_sizes))
    for i, box_size in enumerate(box_sizes):
        # Add a small value to the box size to avoid dividing by zero
        box_size_eps = box_size + 1e-10
        box_counts[i] = np.sum(luminance > box_size_eps)
    # Calculate the logarithm of the box sizes and box counts
    log_box_sizes = np.log(box_sizes + 1e-10)
    log_box_counts = np.log(box_counts + 1e-10)
    # Calculate the slope of the linear regression
    slope, _, _, _, _ = linregress(log_box_sizes, log_box_counts)
    return slope

# Calculate the fractal dimension
turbulence = fractal_dimension(luminance)
````

````python
from PIL import Image
import numpy as np
from scipy.stats import linregress

def fractal_dimension(luminance):
    box_sizes = np.logspace(0, 3, 10)
    box_counts = np.zeros(len(box_sizes))
    for i, box_size in enumerate(box_sizes):
        box_counts[i] = np.sum(luminance > box_size)
    mask = box_counts > 0
    box_sizes = box_sizes[mask]
    box_counts = box_counts[mask]
    log_box_sizes = np.log(box_sizes)
    log_box_counts = np.log(box_counts)
    print(box_sizes,log_box_sizes,log_box_counts)
    if len(box_sizes) > 1:
        slope, _, _, _, _ = linregress(log_box_sizes, log_box_counts)
    else:
        slope = np.nan
    return slope

img = Image.open('vangogh2.jpg')
img_gray = img.convert('L')
img_array = np.array(img_gray)
luminance = img_array / 255.0
turbulence = fractal_dimension(luminance)

if turbulence < 0.5:
    print("Low turbulence: Smooth, calm, and serene")
elif 0.5 <= turbulence <= 1.5:
    print("Medium turbulence: Moderate texture and pattern, some complexity")
else:
    print("High turbulence: Dynamic, expressive, and chaotic")
````

````python
from PIL import Image
import numpy as np
from scipy.stats import linregress

def fractal_dimension(rgb):
    box_sizes = np.logspace(0, 3, 10)
    box_counts = np.zeros(len(box_sizes))
    for i, box_size in enumerate(box_sizes):
        box_counts[i] = np.sum(np.std(rgb, axis=2) > box_size) / (rgb.shape[0] * rgb.shape[1])
    mask = box_counts > 0
    box_sizes = box_sizes[mask]
    box_counts = box_counts[mask]
    print(box_sizes,box_counts)
    if len(box_sizes) > 1:
        log_box_sizes = np.log(box_sizes + 1e-10)
        log_box_counts = np.log(box_counts + 1e-10)
        slope, _, _, _, _ = linregress(log_box_sizes, log_box_counts)
    else:
        slope = np.nan
    return slope

img = Image.open('vangogh2.jpg')
img_array = np.array(img)
rgb = img_array / 255.0
turbulence = fractal_dimension(rgb)

if turbulence < 0.5:
    print("Low turbulence: Smooth, calm, and serene")
elif 0.5 <= turbulence <= 1.5:
    print("Medium turbulence: Moderate texture and pattern, some complexity")
else:
    print("High turbulence: Dynamic, expressive, and chaotic")
````
