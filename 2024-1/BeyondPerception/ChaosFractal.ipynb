import numpy as np
from PIL import Image
import io
import requests
from matplotlib import pyplot as plt

def simulate_deer_vision(image_array):
    # Convert to float for calculations
    img_float = image_array.astype(float) / 255.0
    
    # Extract RGB channels
    r, g, b = img_float[:, :, 0], img_float[:, :, 1], img_float[:, :, 2]
    
    # Simulate dichromatic vision (blue and yellow)
    # These weights are approximations based on general understanding of deer vision
    yellow = 0.5 * r + 0.5 * g  # Combining red and green to simulate yellow sensitivity
    
    # Create new image array with blue and yellow channels
    deer_vision = np.zeros_like(img_float)
    deer_vision[:, :, 0] = yellow  # Use yellow for red channel
    deer_vision[:, :, 1] = yellow  # Use yellow for green channel
    deer_vision[:, :, 2] = b       # Keep blue channel
    
    # Convert back to 8-bit format
    deer_vision = (deer_vision * 255).astype(np.uint8)
    
    return deer_vision

# Function to load and process image
def process_image(image_path):
    # Load image
    img = Image.open(image_path)
    
    # Convert to numpy array
    img_array = np.array(img)
    
    # Simulate deer vision
    deer_vision_array = simulate_deer_vision(img_array)
    
    return img_array, deer_vision_array

# Function to display images side by side
def display_comparison(original, deer_vision):
    plt.figure(figsize=(12, 6))
    
    plt.subplot(121)
    plt.imshow(original)
    plt.title('Original Human Vision')
    plt.axis('off')
    
    plt.subplot(122)
    plt.imshow(deer_vision)
    plt.title('Simulated Deer Vision')
    plt.axis('off')
    
    plt.tight_layout()
    plt.show()

# Main execution
def main():
    # Replace with your image path
    image_path = 'tiger.jpeg'
    
    original, deer_vision = process_image(image_path)
    display_comparison(original, deer_vision)

if __name__ == "__main__":
    main()

import numpy as np
import matplotlib.pyplot as plt

class Boids:
    def __init__(self, Np, Na, home=[0,0]):
        self.Np = Np  # predator
        self.Na = Na  # agent
        self.d = len(home)
        self.home = np.array(home)
        self.predators = np.random.random((Np, self.d*2))
        self.agents = np.random.random((Na, self.d*2))
        self.radius_repulsion = 1
        self.radius_alignment = 0.1
        
    def _pairwise_distances(self, X, Y):
        D = -2 * X @ Y.T + np.sum(Y ** 2, axis=1) + np.sum(X ** 2, axis=1)[:, np.newaxis]
        D[D < 0] = 0
        return D
        
    def _force(self):
        Daa = self._pairwise_distances(self.agents[:,:self.d], self.agents[:,:self.d])
        Dap = self._pairwise_distances(self.agents[:,:self.d], self.predators[:,:self.d])

        Neibors_repulsion = (Daa < self.radius_repulsion).astype(int)
        Neibors_alignment = (Daa < self.radius_alignment).astype(int)
        
        Dr = (1/(Daa+np.eye(self.Na))) * Neibors_repulsion
        Fr = np.diag(np.sum(Dr, axis=1))@ self.agents[:,:self.d] - Dr @ self.agents[:,:self.d]

        Fh = self.home - self.agents[:,:self.d]
    
        Fa = Neibors_alignment @ self.agents[:,self.d:] - 2*self.agents[:,self.d:]
        
        s = 100
        Ff = -np.diag(np.sqrt(np.sum(np.square(self.agents[:,self.d:]), axis=1))- s)/s @ self.agents[:,self.d:]
        
        F_total = Fr + 5*Fh + Ff + Fa
        return 200 * np.tanh(0.1*F_total)
        
    def evolve(self, dt = 0.01):
        self.agents[:,self.d:] += dt*self._force()
        self.agents[:,:self.d] += dt*self.agents[:,self.d:]
        
Na = 30
d = 2
n_history = 10
film = np.zeros((Na, d*n_history))
boids = Boids(0, Na)    
for t in range(100):        
    boids.evolve()
    film = np.hstack([boids.agents[:,:d], film[:,:-d]])
    if t > n_history:
        plt.figure()
        for i in range(Na):
            plt.plot(film[i,0],film[i,1], 'ko')
            plt.plot(film[i,::2],film[i,1::2], 'k-',alpha=0.5)
        bound = 3
        plt.xlim([-bound,bound])
        plt.ylim([-bound,bound])
        plt.title('t={}'.format(t))
        plt.savefig('./fig/{}.jpg'.format(t))
