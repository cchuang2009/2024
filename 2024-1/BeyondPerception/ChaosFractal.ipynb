import numpy as np
from PIL import Image
import io
import requests
from matplotlib import pyplot as plt

def simulate_deer_vision(image_array):
    # Convert to float for calculations
    img_float = image_array.astype(float) / 255.0
    
    # Extract RGB channels
    r, g, b = img_float[:, :, 0], img_float[:, :, 1], img_float[:, :, 2]
    
    # Simulate dichromatic vision (blue and yellow)
    # These weights are approximations based on general understanding of deer vision
    yellow = 0.5 * r + 0.5 * g  # Combining red and green to simulate yellow sensitivity
    
    # Create new image array with blue and yellow channels
    deer_vision = np.zeros_like(img_float)
    deer_vision[:, :, 0] = yellow  # Use yellow for red channel
    deer_vision[:, :, 1] = yellow  # Use yellow for green channel
    deer_vision[:, :, 2] = b       # Keep blue channel
    
    # Convert back to 8-bit format
    deer_vision = (deer_vision * 255).astype(np.uint8)
    
    return deer_vision

# Function to load and process image
def process_image(image_path):
    # Load image
    img = Image.open(image_path)
    
    # Convert to numpy array
    img_array = np.array(img)
    
    # Simulate deer vision
    deer_vision_array = simulate_deer_vision(img_array)
    
    return img_array, deer_vision_array

# Function to display images side by side
def display_comparison(original, deer_vision):
    plt.figure(figsize=(12, 6))
    
    plt.subplot(121)
    plt.imshow(original)
    plt.title('Original Human Vision')
    plt.axis('off')
    
    plt.subplot(122)
    plt.imshow(deer_vision)
    plt.title('Simulated Deer Vision')
    plt.axis('off')
    
    plt.tight_layout()
    plt.show()

# Main execution
def main():
    # Replace with your image path
    image_path = 'tiger.jpeg'
    
    original, deer_vision = process_image(image_path)
    display_comparison(original, deer_vision)

if __name__ == "__main__":
    main()
