````python
def webp2png(webp_image):
    fp = open(webp_image,"rb")
    png_image = PIL.Image.open(fp)
    fp.close()
    png_image.resize((400,400))
    png_image.save(webp_image.replace(".webp", ".png"))
    return png_image
````
